```python
from bs4 import BeautifulSoup
```

```python
import pandas as pd
import requests
url ='https://www.goodreads.com/quotes/tag/friendship'
html_friendship = requests.get(url).text
soup =  BeautifulSoup(html_friendship,'html.parser')
```

```python
stud='https://www.goodreads.com/quotes/tag/friendship?page={}'
```

```python
stud.format(3)
```

```python
def quotes_friendship(soup):
    author_name = [i.text.strip() for i in soup.find_all('span',class_='authorOrTitle')]
    quotes = [list(i)[0].strip() for i in soup.find_all('div',class_='quoteText')]
    likes = [int(i.text.strip().split()[0]) for i in soup.find_all('a',class_='smallText')]
    tags = [i.text.split()[1:] for i in [i for i in soup.find_all('div',class_='greyText smallText left')]]
    return pd.DataFrame({"Authors":author_name,'Quotes':quotes,"Likes":likes,"tags":tags})
```

```python
df = quotes_friendship(soup)
```

```python
table=[]
from tqdm import tqdm
for i in tqdm(range(1,11)):
    url_all= stud.format(i)
    html_all=requests.get(url_all).text
    soup1=BeautifulSoup(html_all,'html.parser')
    table.append(quotes_friendship(soup1))
```

```python
data1=pd.concat(table,axis=0)
data1
```

```python
data1.to_csv("Friendship_Quotes.csv")
```
